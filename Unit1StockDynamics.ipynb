```python
import os,sys
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use("fivethirtyeight")
```

```python
#Load datasets
IBM=pd.read_csv("data/IBMStock.csv")
GE=pd.read_csv("data/GEStock.csv")
ProcterGamble=pd.read_csv("data/ProcterGambleStock.csv")
CocaCola=pd.read_csv("data/CocaColaStock.csv")
Boeing=pd.read_csv("data/BoeingStock.csv")
```

```python
#length of data sets
len(IBM)
```

```python
#Set dates from factors to datetime indices
stocks=[IBM,GE,ProcterGamble,CocaCola,Boeing]
for i in stocks:
    i.set_index(pd.DatetimeIndex(i['Date']),inplace=True)
    i.drop("Date",axis=1,inplace=True)
    
IBM.columns=["IBM"]
GE.columns=["GE"]
ProcterGamble.columns=["PG"]
CocaCola.columns=["CC"]
Boeing.columns=["Boeing"]

    
```

```python
#Earliest year
Boeing.index.min().year
```

```python
#Latest year
Boeing.index.max().year
```

```python
#Minimum price of GE over this time period
GE.min()
```

```python
#Maximum price of Coca-Cola over this period
CocaCola.max()
```

```python
#Median price of Boeing
Boeing.median()
```

```python
#Standard deviation of the stock price of Procter & Gamble over this period
ProcterGamble.std()
```

```python
CocaCola.plot(style=":")
```

```python
#Visualizing stock dynamics
fig=plt.figure(figsize=[16,10])
ax=plt.axes()
ax.plot(CocaCola['CC'][301:432],":")
ax.plot(ProcterGamble['PG'][301:432])
ax.plot(IBM['IBM'][301:432])
ax.plot(GE["GE"][301:432])
ax.plot(Boeing["Boeing"][301:432])
ax.legend()
```

```python
#Which months did IBM have historically the highest stock price?
IBM.groupby(IBM.index.month).mean().sort_values("IBM",ascending=False)[:5]
```

```python
#Which months did GE have historically the highest stock price?
GE.groupby(GE.index.month).mean().sort_values("GE",ascending=False)[:5]
```

```python
#Which months did Coca-cola have the highest stock price?
CocaCola.groupby(CocaCola.index.month).mean().sort_values("CC",ascending=False)[:5]
```
