```python
from sklearn.linear_model import LinearRegression

pTrain=pd.read_csv('data/pisa2009Train.csv')
pTest=pd.read_csv('data/pisa2009Test.csv')

pTrain=pTrain.dropna()
pTest=pTest.dropna()

trainDummies=pd.get_dummies(pTrain['raceeth'])
pTrain=pd.concat([pTrain,trainDummies],axis=1)
pTrain.drop('raceeth',axis=1,inplace=True)
testDummies=pd.get_dummies(pTest['raceeth'])
pTest=pd.concat([pTest,testDummies],axis=1)
pTest.drop('raceeth',axis=1,inplace=True)


xTrain=pTrain.drop('readingScore',axis=1)
yTrain=pTrain['readingScore']
xTrain=sm.add_constant(xTrain)

xTest=pTest.drop('readingScore',axis=1)
yTest=pTest['readingScore']
xTest=sm.add_constant(xTest)

model=sm.OLS(yTrain,xTrain.astype(float)).fit()

model.summary()
```

```python
yPred=model.predict(xTest)

math.sqrt(mse(yTest,yPred))

sst=mse([yTest.mean()]*len(yPred),yTest)*len(yPred)
sse=mse(yTest,yPred)*len(yPred)
1-(sse/sst)
```
